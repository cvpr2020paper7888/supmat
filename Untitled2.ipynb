```python
% matplotlib inline
import torch
import torchvision
```

```python
from matplotlib import pyplot as plt
```

```python
from task_configs import GaussianBulr as GaussianBlur
from task_configs import tasks
from datasets import ImageDataset
```

```python
SHARED_DIR = 'mount/shared'
```

```python
batch_size = 30
n_workers = 8
frame_dir = f"{SHARED_DIR}/assets/input_frames"
frame_loader = torch.utils.data.DataLoader(
    ImageDataset(tasks=[tasks.rgb(size=256)], data_dir=frame_dir),
    batch_size=batch_size,
    num_workers=n_workers, shuffle=False, pin_memory=True
)
```

```python
topil = torchvision.transforms.ToPILImage()
totensor = torchvision.transforms.ToTensor()
```

```python
frame_dir = f"{SHARED_DIR}/assets/input_frames"
output_dir = f"{SHARED_DIR}/assets/input_frames_blurred"

counter = 0
blur_radii = [0, 1, 2, 3, 4, 5]
for i, frames in enumerate(frame_loader):
    counter += 1
    blur = GaussianBlur(blur_radii[i % len(blur_radii)])
    for frame in frames[0]:
        im = blur(topil(frame))
        im.save(f"{output_dir}/blurred{counter:05}.png","PNG")
    if i == 4: 
        break
```

```python
outout_frame_loader = torch.utils.data.DataLoader(
    ImageDataset(tasks=[tasks.rgb(size=256)], data_dir=output_dir),
    batch_size=batch_size,
    num_workers=n_workers, shuffle=False, pin_memory=True
)
```
